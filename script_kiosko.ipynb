```python
import openpyxl
import numpy as np
import pandas as pd
```

```python
route=r'd:\Descargas\reporte1.xlsx'
df=pd.read_excel(route, sheet_name=0)
df2=pd.read_excel(route,sheet_name=1)
df2=pd.read_excel(route,sheet_name=2)
```

```python
df.head()
```

```python
df['Total'] = pd.to_numeric(df['Total'].str.replace(',', '', regex=False), errors='coerce')
df['Subtotal'] = pd.to_numeric(df['Subtotal'].str.replace(',', '', regex=False), errors='coerce')
```

```python
df_totales=df[['Total','Subtotal']]
resumen_personalizado = pd.DataFrame({
    'count': df_totales.count(),
    'mean': df_totales.mean(),
    'min': df_totales.min(),
    'max': df_totales.max(),
    'suma': df_totales.sum()
})
resumen_personalizado
```

```python
df['Fecha']=pd.to_datetime(df['Fecha'])
resumen_por_dia = df.groupby(df['Fecha'].dt.date).agg(
    suma=('Total', 'sum'),
    maximo=('Total', 'max'),
    minimo=('Total', 'min')
).reset_index()
resumen_por_dia
```

```python
resumen_por_semana = df.groupby(df['Fecha'].dt.isocalendar().week).agg(
    suma=('Total', 'sum'),
    maximo=('Total', 'max'),
    minimo=('Total', 'min')
).reset_index()
resumen_por_semana
```
